```python
date_columns = ['opendate', 'ldrdate', 'lcrdate', 'dob']
for col in date_columns:
    # Convert to datetime while handling errors
    processed_df[col] = pd.to_datetime(processed_df[col], format='%d-%b-%y', errors='coerce')
        
```

```python
from datetime import datetime
import numpy as np
current_date = datetime.now()
```

```python
processed_df
```

```python
processed_df['is_personal'] = processed_df['sector'].isin(['1500','1510','2000']) & processed_df['industry'].isin(['9009','9008'])
```

```python
processed_df
```

```python
if processed_df['is_personal']:
    if processed_df['dob'] is not None:
        processed_df['age'] = np.abs(processed_df['age'].fillna(0)) + processed_df['account_age_years']
```

```python
processed_df
```

```python
processed_df['age'] = np.where(
            processed_df['is_personal'] & 
            processed_df['dob'].notna(),
            (current_date - processed_df['dob']).dt.days // 365,
            None
        )
```

```python
processed_df['age'].describe()
```

```python
processed_df['sector'].value_counts()
```

```python
# Calculate account age for all accounts
processed_df['account_age_years'] = np.where(
    processed_df['opendate'].notna(),
    (current_date - processed_df['opendate']).dt.days // 365,
    None
)
```

```python
system_message = """
You're financial expert with professional knowledge about Python and Pandas. 
You have been asked to analyze the financial data of a company.
The company resides in Nepal and the currency used is Nepalese Rupee (NPR).
You have been provided with the following data.
"""

# prompt = system_message
# prompt = system_message + data_json
prompt = system_message + data_json + "\n What insights can you provide from the data?" # + explanation
```

```python
result = ollama.generate(model=MODEL, prompt=prompt)
print(result.done)

if result.done:
    print(result.response)
else:
    print(result.model + " .Please try again later.")
```
